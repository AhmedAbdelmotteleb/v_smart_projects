Script that demonstrates Collatz conjecture:

Take any natural number n.

If n is even, divide it by 2 to get n / 2.

If n is odd, multiply it by 3 and add 1 to get 3n + 1.

Repeat the process indefinitely.

The conjecture states that no matter which number you start with, you will always reach 1 eventually.

This script also demonstrates this visually.

```python
import matplotlib.pyplot as plt

def collatz(n):
    """
    Function that takes a number n and returns the next number in the Collatz sequence
    """
    if n % 2 == 0:
        return n // 2
    else:
        return 3*n + 1

def collatz_sequence(n):
    """
    Function that takes a number n and returns the Collatz sequence starting at n
    """
    sequence = [n]
    while n != 1:
        n = collatz(n)
        sequence.append(n)
    return sequence

def collatz_plot(n):
    """
    Function that plots the Collatz sequence starting at n
    """
    sequence = collatz_sequence(n)
    plt.plot(sequence)
    plt.xlabel('Iteration')
    plt.ylabel('Value')
    plt.title(f'Collatz sequence starting at {n}')

    # Highlight the end of the sequence
    plt.plot(len(sequence) - 1, sequence[-1], 'ro', label=len(sequence)-1) 
    plt.legend()
    
    plt.show()

def collatz_plot_multiple(n, m):
    """
    Function that plots the Collatz sequences ranging from n to m
    """
    for i in range(n, m+1):
        collatz_plot(i)


collatz_plot(50)
collatz_plot_multiple(20, 23)
```
